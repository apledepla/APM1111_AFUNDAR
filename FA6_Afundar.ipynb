```python
from IPython.display import display
import pandas as pd

N = 120
q3 = (3 * N) / 4

lb = 90
freq = 9
cumfreq = 120
q3_ans = lb + ((q3 - cumfreq) / freq)

q1 = N / 4
lb = 30
freq = 1
cumfreq = 1
q1_ans = lb + ((q1 - cumfreq) / freq)

q2 = (2 * N) / 4
lb = 60
freq = 21
cumfreq = 36
q2_ans = lb + ((q2 - cumfreq) / freq)

ansdf = pd.DataFrame({
    'Quartile 1': [q1_ans],
    'Quartile 2': [q2_ans],
    'Quartile 3': [q3_ans]
})

display(ansdf)
```

```python
stats_sd=8
stats_m=78

alge_sd=7.6
alge_m=73

relative_disp_stats= (stats_sd/stats_m)*100
relative_disp_alge= (alge_sd/alge_m)*100

ansdf = pd.DataFrame({
    'Relative Dispersion For Statistics': [relative_disp_stats],
    'Relative Dispersion For Algebra': [relative_disp_alge],
    'Absolute Dispersion For Statistics': [stats_sd],
    'Absolute Dispersion For Algebra': [alge_sd]
})

display(ansdf)

print("(a) The absolute dispersion for Statistics is greater than the Algebra while (b) the relative dispersion for algebra is greater than the Statistics")
```

```python
import math

set3 = [6,2,8,7,5]
mean = sum(set3)/len(set3)
sd = math.sqrt(sum((x-mean)**2 for x in set3)/len(set3))
z=[]

for i in set3:
   z.append((i-mean)/sd)

meanz = sum(z)/len(z)
sdz = math.sqrt(sum((x-meanz)**2 for x in z)/len(z))

print(f'The z scores are: {z}')
print(f'The mean of the z scores are: {meanz}')
print(f'The Standard Deviation of the z scores are: {sdz}')
print("Therefore, The mean and standard deviation of the z-scores are between 0 and 1")
```

```python
setx = [6, 9, 12, 15, 18]
px = [0.1, 0.2, 0.4, 0.2, 0.1]

meanx = sum(x * p for x, p in zip(setx, px))
variancex = sum(((x - meanx) ** 2) * p for x, p in zip(setx, px))

print(f"Mean: {meanx:.2f}")
print(f"Variance: {variancex}")

setForSets = []
meanForSets = []
probSets = []

for x in range(len(setx)):
    for y in range(len(setx)):
        setholder = [setx[x], setx[y]]
        meanForSets.append((setx[x] + setx[y]) / 2)
        probSets.append((px[x] + px[y]) / 10)
        setForSets.append(setholder)


print("\nGrouped Results:")
for i in range(len(setForSets)):
    print(f"Set: {setForSets[i]}")
    print(f"Mean: {meanForSets[i]}, Probability: {probSets[i]:.2f}\n\n")
```
